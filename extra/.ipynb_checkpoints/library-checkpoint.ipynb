```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_path = "C:\\Users\\ozcan\\Downloads\\library_koca.jpg"
img = cv.imread(image_path)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img)
plt.show()
```

```python
denoised_img = cv.fastNlMeansDenoising(gray_img, None, h=10, searchWindowSize = 21, templateWindowSize = 7)
```

```python
plt.imshow(denoised_img)
plt.show()
```

```python
edges = cv.Canny(denoised_img, 50, 100)
sharpening_filter = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1,-1,-1]])
sharpened_img = cv.filter2D(denoised_img, -1, sharpening_filter)
colorized_img = cv.cvtColor(sharpened_img, cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10, 7))

plt.subplot(231), plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(232), plt.imshow(gray_img, cmap='gray'), plt.title('Grayscale Image')
plt.subplot(233), plt.imshow(denoised_img, cmap='gray'), plt.title('Denoised Image')
plt.subplot(234), plt.imshow(edges, cmap='gray'), plt.title('Edge Detection')
plt.subplot(235), plt.imshow(sharpened_img, cmap='gray'), plt.title('Sharpened Image')
plt.subplot(235), plt.imshow(colorized_img), plt.title('Colorized Image')

plt.show()
```

```python
cv.imwrite('colorized_libkoca1.jpg', colorized_img)
```
