```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img_path = "C:\\Users\\ozcan\\pexels-benjamin-suter-3733269.jpg"
img = cv2.imread(img_path)
print(img.shape)
```

```python
gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.show()
print(gray_image.shape)
```

```python
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(blackAndWhiteImage)
plt.show()
```

```python
output2 = cv2.blur(gray_image, (30,30))
plt.imshow(output2)
plt.show()
```

```python
output2 = cv2.GaussianBlur(gray_image,(19,19) ,10)
plt.imshow(output2)
plt.show()
```

```python
(h,w) = img.shape[:2]
center = (w/2, h/2)
M = cv2.getRotationMatrix2D(center,13,scale =1.1)
rotated = cv2.warpAffine(gray_image,M,(w,h))
plt.imshow(rotated)
plt.show()
```
